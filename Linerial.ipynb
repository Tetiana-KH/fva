```python
import statsmodels.formula.api as smf
import pandas as pd
from matplotlib import pyplot as plt
```

```python
fvaFile = pd.read_csv("c:/fva/#data#/fvaFile.csv")
#fvaFile.head()
```

```python
model = smf.ols('ID ~ DevId', data=fvaFile)
model = model.fit()
```

```python
# Predict values
devId_pred = model.predict()

# Plot regression against actual data
plt.figure(figsize=(12, 6))
plt.plot(fvaFile['DevId'].to_numpy(), fvaFile['ID'].to_numpy(), 'o')
# scatter plot showing actual data
plt.plot(fvaFile['DevId'].to_numpy(), devId_pred, 'r', linewidth=2)
#regression line
plt.xlabel('DevId')
plt.ylabel('ID')
plt.title('DevId vs ID')
plt.show()
```
